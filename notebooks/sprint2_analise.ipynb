```python
import pandas as pd
import sqlite3

# 1. Conectando ao banco de dados da Sprint 1 (voltamos uma pasta com ../)
conexao = sqlite3.connect('../data/processed/ecommerce.db')

# 2. Puxando a tabela inteira usando uma consulta SQL básica
df_vendas = pd.read_sql_query("SELECT * FROM vendas", conexao)

# 3. Exibindo as 5 primeiras linhas para garantir que os dados vieram certinhos
display(df_vendas.head())
```

```python
# PERGUNTA 1: Qual o faturamento total da loja?
faturamento = df_vendas['Valor_Total'].sum()
print(f"Faturamento Total da Loja: R$ {faturamento:,.2f}\n")

# PERGUNTA 2: Quais são os 3 produtos mais vendidos (em quantidade)?
produtos_top3 = df_vendas.groupby('Nome_Produto')['Quantidade'].sum().sort_values(ascending=False).head(3)
print("--- TOP 3 Produtos Mais Vendidos ---")
print(produtos_top3, "\n")

# PERGUNTA 3: Qual categoria traz mais receita? (Resolvendo com SQL avançado)
query_categorias = """
    SELECT Categoria_Produto, SUM(Valor_Total) as Receita_Total
    FROM vendas
    GROUP BY Categoria_Produto
    ORDER BY Receita_Total DESC
"""
df_categorias = pd.read_sql_query(query_categorias, conexao)

print("--- Receita por Categoria (Consulta SQL) ---")
display(df_categorias)

# Fechando a conexão com o banco por segurança
conexao.close()
```
